```python
import sys,pprint
```

```python
def createCalendar (people : list, prices : list, n : int): #n = number of people
    rounded_prices = round_prices(prices)
    prices_sum = sum(rounded_prices)
    total_days = prices_sum*2
    days_per_person = [(2*price) for price in rounded_prices]
    ratios = [(day/total_days) for day in days_per_person]
    actual_ratios = [(price/sum(prices)) for price in prices]
    diff = [(total_days*(ratio-actual_ratio)) for (ratio,actual_ratio) in zip(ratios,actual_ratios)]
    schedule = []
    days_left = days_per_person.copy()
    d = 0
    for i in range(0, int(total_days)):
        if d==7: d=0
        while(days_left[d] <= 0):
            if d==7: d=0
            d += 1
        schedule.append(people[d])
        days_left[d] -= 1
        d += 1
    return schedule, days_per_person, diff
```

```python
def roundPrices(prices : list):
    rounded_prices = []
    for price in prices:
        dec = price % 1
        if dec <= 0.25:
            rounded_prices.append(price-dec)
        elif dec > 0.25 and dec <=0.75:
            rounded_prices.append(price-dec + 0.50)
        else:
            rounded_prices.append(price-dec+1)
    return rounded_prices
```

```python
people = ['Bob', 'Jeremy', 'Sunny', 'Allison', 'Julie', 'Sophie', 'Angela']
prices = [5.67,4.23,7.86,6.65,6.30,4.78,9.25]
n = 7
pprint.pprint(createCalendar(people, prices,n))
```
